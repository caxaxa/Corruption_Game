```python

import numpy as np
import matplotlib.pyplot as plt 



# Indiference Curves
n = 100
x = np.arange(0.0000001, 1,1/n )
beta = []
alpha = []
b = 4
a = 6
f = 10
N = 0.5
c = 2
gamma = 0.9

alphaIE = []
for i in x:
    alphaIE.append((-(b/(gamma**2)) + a)/ (i*(a+f))) #

alphaIE = np.asarray(alphaIE)      
    
alphaIB = []

for i in x:
    alphaIB.append((-(c/(gamma**2)) + b)/ (i*(b+f)))  #
    
alphaIB = np.asarray(alphaIB)   

#Ploting Indiference curves

fig, ax = plt.subplots(figsize = (5,4))

ax.plot(x, alphaIE, linewidth=2)
ax.plot(x, alphaIB,'--', linewidth=2)

fig.add_subplot(ax)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

fig.text(0.92, 0.05, '$𝛽$')
fig.text(0.11, 0.9, '$𝛼$')

ax.fill_between(x, alphaIE, facecolor='salmon', alpha=0.1)

plt.ylim([0, 1])
plt.xlim([0, 1])
plt.xticks(np.arange(0,1.1,1))
plt.yticks(np.arange(0,1.1,1))
#plt.title('Indiference curves')
plt.legend(('IC Payer', 'IC Receiver', 'Corruption Zone' ),
           loc='best', shadow=False)

# fig.text(0.18, 0.2, 'Corruption')
# fig.text(0.5, 0.4, 'No Corruption')
plt.show()

fig.savefig('fig1.pdf')
```

```python
k = (b-a)/(b**2)

R_star = 1.12

R = []
for i in x:
    R.append((R_star+k-1)/(i*k)) #
    
R = np.asarray(R)  

P_star = 6.3

P = [P_star/((k-1)/k)]*n

P = np.asarray(P)  
    
fig, ax = plt.subplots(figsize = (5,4))

ax.plot(x, alphaIE, linewidth=2)
ax.plot(x, alphaIB,'--', linewidth=2)
ax.plot(x,R,'-.',linewidth=.5 )
ax.plot(P, x,linewidth=.5 )

fig.add_subplot(ax)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

fig.text(0.92, 0.05, '$𝛽$')
fig.text(0.1, 0.9, '$𝛼$')
fig.text(0.65, 0.08, '$𝛽~(P*) $')
#fig.text(0.02, 0.485, '$𝛼~(R*)$')

ax.fill_between(x[0:71], alphaIE[0:71], facecolor='salmon', alpha=0.1)
ax.fill_between(x[70:100], alphaIE[70:100], facecolor='green', alpha=0.1)
ax.fill_between( x, alphaIE , R ,facecolor='white', alpha=1,label='_nolegend_')
ax.fill_between(x, alphaIE, R, facecolor='yellow', alpha=0.1)

plt.ylim([0, 1])
plt.xlim([0, 1])
plt.xticks(np.arange(0,1.1,1))
plt.yticks(np.arange(0,1.1,1))
#plt.title('Indiference curves')
plt.legend(('IC Payer', 'IC Receiver','Self-Reporting Bonus','Plea Agreement', 'Corruption Zone','Plea Bragaining','Self-Reporting'),
           loc='upper right', shadow=False)

plt.show()

fig.savefig('fig2.pdf')
```

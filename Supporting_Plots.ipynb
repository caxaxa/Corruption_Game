```python

import numpy as np
import matplotlib.pyplot as plt 



# Indiference Curves
n = 100
x = np.arange(0.0000001, 1,1/n )
beta = 0.70
alpha = 0.05
b = 3
a = 6
f = 12
N = 0.5
c_b = 1
gamma = 0.99

def get_ICs():
    n = 100
    x = np.arange(0.0000001, 1,1/n )
    N = 0.5

    alphaIE = []
    for i in x:
        alphaIE.append((-(b/(gamma**2)) + a)/ (i*(a+f))) #

    alphaIE = np.asarray(alphaIE)      

    alphaIB = []

    for i in x:
        alphaIB.append((-(c_b/(gamma**2)) + b)/ (i*(b+f)))  #

    alphaIB = np.asarray(alphaIB)   

    #Ploting Indiference curves

    fig, ax = plt.subplots(figsize = (5,4))

    ax.plot(x, alphaIE, linewidth=2)
    ax.plot(x, alphaIB,'--', linewidth=2)

    fig.add_subplot(ax)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    fig.text(0.92, 0.05, '$ùõΩ$')
    fig.text(0.11, 0.9, '$ùõº$')

    ax.fill_between(x, alphaIE, facecolor='salmon', alpha=0.1)

    plt.ylim([0, 1])
    plt.xlim([0, 1])
    plt.grid(True)
    plt.xticks(np.arange(0,1.1,.1))
    plt.yticks(np.arange(0,1.1,.1))
    #plt.title('Indiference curves')
    plt.legend(('IC Payer', 'IC Receiver', 'Corruption Zone' ),
               loc='best', shadow=False)
    #plt.grid(True)
    # fig.text(0.18, 0.2, 'Corruption')
    # fig.text(0.5, 0.4, 'No Corruption')
    plt.show()
get_ICs()
fig.savefig('fig1.pdf')
```

```python
R_star = -(gamma**2)*((1-(alpha*beta))*a -alpha*beta*f)/f
P_star = -gamma*((1-beta)*b -beta*f)/f
print(R_star)
print(P_star)
```

```python

def get_IC_star(R_star,P_star):

    R = []
    for i in x:
        R.append((((gamma**2)*a) +( R_star*f))/(((gamma**2)*(a+f))*i)) #

    R = np.asarray(R)  


    P = [((gamma*a) +( P_star*f))/(gamma*(a+f))]*n

    lim = int(((gamma*a) +( P_star*f))/(gamma*(a+f))*100)

    P = np.asarray(P)  

    fig, ax = plt.subplots(figsize = (5,4))

    ax.plot(x, alphaIE, linewidth=2)
    ax.plot(x, alphaIB,'--', linewidth=2)
    ax.plot(x,R,'-.',linewidth=.5 )
    ax.plot(P, x,linewidth=.5 )

    fig.add_subplot(ax)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    fig.text(0.92, 0.05, '$ùõΩ$')
    fig.text(0.1, 0.9, '$ùõº$')
    #fig.text(0.65, 0.08, '$ùõΩ~(P*) $')
    #fig.text(0.02, 0.485, '$ùõº~(R*)$')

    ax.fill_between(x[0:lim + 1], alphaIB[0:lim + 1], facecolor='salmon', alpha=0.1)
    ax.fill_between(x[lim:100], alphaIB[lim:100], facecolor='green', alpha=0.1)
    ax.fill_between( x, alphaIB , R ,facecolor='white', alpha=1,label='_nolegend_')
    ax.fill_between(x, alphaIB, R, facecolor='yellow', alpha=0.1)

    plt.ylim([0, 1])
    plt.xlim([0, 1])
    plt.grid(True)
    plt.xticks(np.arange(0,1.1,.1))
    plt.yticks(np.arange(0,1.1,.1))
    #plt.title('Indiference curves')
    plt.legend(('IC Payer', 'IC Receiver','Self-Reporting Bonus','Plea Agreement', 'Corruption Zone','Plea Bragaining','Self-Reporting'),
               loc='upper right', shadow=False)

    plt.show()
    
get_IC_star(R_star,P_star)

fig.savefig('fig2.pdf')
```
